```python
# Load matplotlib and make a plot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Making a basic figure with matplotlib
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python

# Making a histogram 
fig,ax=plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Customizing plot labels
fig,ax = plt.subplots(figsize=(5,3))
ax. plot(x,y)

ax.set_title('Totel growth over time')
ax.set_ylabel('Total growth(%)')
ax.set_xlabel('Years since start')
```

```python
# dictionary for the labels
font1 = {'family': 'sans-serif','color':'blue', 'size': 20}
font2 = {'family':'monospace','color':'green','size': 15 }

fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time', fontdict = font1, loc='right')
ax.set_ylabel('Total grown', fontdict = font2, loc='bottom')
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# Customizing plot appearnace

fig,ax = plt.subplots(figsize = (5,3))
ax.scatter(x,
           y,
           marker = 'o',
           color = 'red')
```

```python
# Modifying data points

fig,ax = plt.subplots(figsize=(5,3))
ax. plot(x,
         y,
         marker = '*',
         color = '#7425b9',
         linestyle = '-',
         linewidth = '2')
```

```python
#Customersize the marker
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 8,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'x', color = 'pink', linestyle = '--', linewidth = 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# try to plot both y variables on the same axis

fig,ax = plt.subplots(figsize=(5,3))
ax. plot(x,y1)
ax. plot(x,y2)
```

```python
# Adding a legend
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label = 'Person 1')
ax.plot(x, y2, label = 'Person 2')
ax. legend(loc = 'lower right',
           frameon = True,
           fontsize = 12,
           ncol = 2,
           shadow = True)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label ='Person 1')
ax.plot(x,y2,label = 'Person 2')
ax.legend(loc = 'upper right',
          bbox_to_anchor=(1,1))
```

```python
# Annotations with the text() function
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = 'Person 1')
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95, "Some random words!",
        ha = 'center',
        color ='red',
        size =20)
```

```python
fig,ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5,"Data: (1,5)", transform = ax.transData)
ax.text(0.5, 0.1, "Axes: (0.5,0.1)", transform = ax.transAxes)
ax.text(0.2, 0.2,"Figure: (0.2, 0.2)", transform = fig.transFigure)
```

```python
fig,ax = plt.subplots(figsize = (5,3))
ax. scatter(x,y1,label = 'Person 1')
ax. scatter(x,y2, label = 'Person 2')
ax.legend(loc = 'lower right')

ax.annotate("This is important!", xy=(10,95), xytext=(20,94),
            arrowprops = dict(arrowstyle = 'wedge',
                              color = 'hotpink',))
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax. legend(loc = 'lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```
