```python
# Load matplotlib and make a plot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Making a basic figure with matplotlib
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python

# Making a histogram 
fig,ax=plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Customizing plot labels
fig,ax = plt.subplots(figsize=(5,3))
ax. plot(x,y)

ax.set_title('Totel growth over time')
ax.set_ylabel('Total growth(%)')
ax.set_xlabel('Years since start')
```

```python
# dictionary for the labels
font1 = {'family': 'sans-serif','color':'blue', 'size': 20}
font2 = {'family':'monospace','color':'green','size': 15 }

fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time', fontdict = font1, loc='right')
ax.set_ylabel('Total grown', fontdict = font2, loc='bottom')
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# Customizing plot appearnace

fig,ax = plt.subplots(figsize = (5,3))
ax.scatter(x,
           y,
           marker = 'o',
           color = 'red')
```

```python
# Modifying data points

fig,ax = plt.subplots(figsize=(5,3))
ax. plot(x,
         y,
         marker = '*',
         color = '#7425b9',
         linestyle = '-',
         linewidth = '2')
```

```python
#Customersize the marker
fig,ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 8,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'x', color = 'pink', linestyle = '--', linewidth = 3)
```
